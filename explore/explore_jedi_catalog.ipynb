Altair can't handle special characters in column names (e.g., ., [], %) without first escaping them with \\

```python
jedi_sub = jedi[['19.5 Depth [%]', '17.1 Depth [%]']]
```

```python
alt.Chart(jedi_sub).mark_point().encode(
    x='19\\.5 Depth \\[\\%\\]:Q',
    y='17\\.1 Depth \\[\\%\\]:Q'
)
```

# Try out a Principal Component Analysis for dimensionality reduction
There are 24k columns in JEDI... what does that look like if I reduce it to two

```python
# Clean the data -- PCA can't handle string types so I have to drop all of my timestamps, which really I should drop anyway for this analysis
jedi_no_times = jedi.select_dtypes(exclude=['object'])

# PCA also can't handle NaN so I need to impute (fill) those values with something
```

```python
model = PCA(n_components=2)

model = make_pipeline(Imputer(strategy='mean'),
                      PCA(n_components=2))
model.fit(jedi_no_times)
jedi_2d = model.transform(jedi_no_times)
```

```python
jedi_no_times['PCA1'] = jedi_2d[:, 0]
jedi_no_times['PCA2'] = jedi_2d[:, 1]
sns.lmplot("PCA1", "PCA2", data=jedi_no_times)
```

```python
plt.scatter(jedi_2d[:, 0], jedi_2d[:, 1], alpha=0.2)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2");
```
